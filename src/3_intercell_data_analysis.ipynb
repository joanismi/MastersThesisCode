# Clean data and filter interactions
There are interactions in the retrieved intercellular interactions network that are not truly intercellular interactions. For example, interactions involving TP53 are in fact transcription factor - target gene interactions.

We will remove this interactions by requiring that proteins have to belong to 1 of the 3 intercellular locations assigned in the network:

* secreted

* plasma_membrane_transmembrane

* plasma_membrane_peripheral

We will also create a **literature curated network** by removing the records with no references. We will use both curated and non-curated networks in our subsequent analysis. The non-curated network has more interactions and proteins but has also more noise, i.e., interactions that might not happen in physiological conditions, for example, interactions caught in high throughput screenings.

```python
print(network.shape)
network_new = network[
    (network.secreted_intercell_source==True)|
    (network.plasma_membrane_transmembrane_intercell_source==True)|
    (network.plasma_membrane_peripheral_intercell_source==True)
]
print(network_new.shape)
network_new = network_new[
    (network_new.secreted_intercell_target==True)|
    (network_new.plasma_membrane_transmembrane_intercell_target==True)|
    (network_new.plasma_membrane_peripheral_intercell_target==True)
]
print(network_new.shape)
```

```python
# Filter and clean data: remove duplicate interactions & entries without gene_symbol
# choose columns with gene symbols (HGNC identifiers)
intercell = network_new[[ 
    'genesymbol_intercell_source',
    'genesymbol_intercell_target',
    'is_stimulation',
    'is_inhibition',
    'consensus_direction']
    ].rename(columns={
        'genesymbol_intercell_source':'source',
        'genesymbol_intercell_target':'target'
    }).dropna().drop_duplicates().reset_index(drop=True)
intercell.shape
```

```python
intercell[intercell.source==intercell.target]
```

```python
# Integrins are obligate heterodimers, formed by alpha and beta subunits
# In the intercellular networks they appear as complexes of 2 genes: the alpha and beta genes
# That means that they are not simple transient complexes of proteins but 1 protein with 2 subunits
# So, by unfolding complexes we are considering interactions that might not exist
index = intercell.target.str.findall('ITG').apply(lambda x: np.nan if len(x)==0 else x[0]).dropna().index
intercell.loc[index].sort_values('source')
```

```python
# create a dataset with only curated interactions
# this interactions have, at least one entry in the column "references"
intercell_curated = network_new.dropna(subset=['references'])[[
    'genesymbol_intercell_source',
    'genesymbol_intercell_target',
    'is_stimulation',
    'is_inhibition',
    'consensus_direction'
]].rename(columns={
        'genesymbol_intercell_source':'source',
        'genesymbol_intercell_target':'target'
    }).dropna().drop_duplicates().reset_index(drop=True)
intercell_curated.shape
```

```python
# some ids are complexes with the format -> COMPLEX:id1_id2... 
# some interactions are from 1 complex to another
intercell[intercell.source.str.contains(r'[:_]') & intercell.target.str.contains(r'[:_]')].head()
```

```python
# unfold all complexes and assign interactions with each complex proteins
# interactions without considering the effect
intercell_simp = id_mapping.unfold_complexes(intercell[['source', 'target']])
intercell_curated_simp = id_mapping.unfold_complexes(intercell_curated[['source', 'target']])

intercell_simp.head()
```

```python
intercell_simp[intercell_simp.source==intercell_simp.target]
```

```python
# Integrins are obligate heterodimers, formed by alpha and beta subunits
# In the intercellular networks they appear as complexes of 2 genes: the alpha and beta genes
# That means that they are not simple transient complexes of proteins but 1 protein with 2 subunits
# So, by unfolding complexes we are considering interactions that might not happen
index = intercell_simp.target.str.findall('ITG').apply(lambda x: np.nan if len(x)==0 else x[0]).dropna().index

intercell_simp.loc[index].sort_values('source')
```

```python
# create new datasets with all interactions considering the effect
# in this dataset, all interactions are either a stimulation or an inhibition
intercell_effect = id_mapping.unfold_complexes(intercell.query('consensus_direction==True'))
intercell_curated_effect = id_mapping.unfold_complexes(intercell_curated.query('consensus_direction==True'))
intercell_effect.head()
```

```python
graphs = {
    'Simple graph': intercell_simp,
    'Simple curated graph': intercell_curated_simp,
    'Effect graph': intercell_effect,
    'Effect curated graph': intercell_curated_effect,
}

for k, v in graphs.items():
    print(k)
    print('{:} {:}'.format('number of interactions:', v.shape[0]))
    print('{:>23} {:}'.format('source genes:', len(v.source.unique())))
    print('{:>23} {:}'.format('target genes:', len(v.target.unique())))
    print('{:>23} {:}'.format('distinct genes:', len(
        set(list(v.source.unique())+list(v.target.unique()))))
         )
    print()
```

```python
# export data
intercell_simp.to_csv(intercell_net_dir + 'intercell_graph.csv', index=False)
intercell_curated_simp.to_csv(intercell_net_dir + 'intercell_curated_graph.csv', index=False)
```
